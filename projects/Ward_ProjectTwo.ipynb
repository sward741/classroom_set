```python
import pandas as pd
foss = pd.read_csv("data/FossilTNRS.csv")
foss
```

## Question 2. Produce a histogram of the minimum ages of the ant specimens in this dataset (15 pts) . The name of the X-Axis should be "Age of Fossil (millions of years)". 

```python
foss.columns = (['taxon', 'reference_no', 'subfamily', 'Tribe', 'Genus', 'Fossil',
       'Age of Fossil (millions of years)', 'max_ma', 'Notes'])
```

```python
import plotnine as p9
(p9.ggplot(foss)  
    + p9.aes(x='Age of Fossil (millions of years)',
            fill = 'subfamily') 
    + p9.geom_histogram(binwidth=8))
```

## Question 3. Produce a bar plot of how many fossils exist for each subfamily. (10 pts). 

```python
my_custom_theme = p9.theme(axis_text_x = p9.element_text(color="blue", size=10,
                                                         angle=90, hjust=.5),
                           axis_text_y = p9.element_text(color="blue", size=12))
(p9.ggplot(data=foss,
           mapping=p9.aes(x='subfamily', fill= 'subfamily'))
    + p9.geom_bar()
    + p9.theme_light()
    + my_custom_theme)
```

# Part Two: Genetic Data

In the data directory, there is an ~unknown~ DNA sequence in a FASTA file. 

## Question 4: Perform a BLAST search of this ~unknown~ sequence (5 pts)

```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO

mystery = SeqIO.read("../project_two/data/mysterSequence-2.fasta", format="fasta")
myst_result = NCBIWWW.qblast("blastn", "nt", mystery.format("fasta"))
```

## Question 5: Create a results folder and write your sequence to it. Not all of this question needs to be completed in python (5 pts)

```python
unknown = myst_result
```

```python
save_file = open("unknownresult", "w")
save_file.write(unknown.read())
save_file.close()
```

## Question 6: What type of organism do you think these data came from? (5pts)

```python
from Bio.Blast import NCBIXML

parsed = NCBIXML.read(open("results/unknownresult"))
alignment_set = parsed.alignments
for individual_alignment in alignment_set:
    print(individual_alignment)
```

I think this sequence came from an ant. The first three results, Aenictogiton, Cerapachys, and Strumigenys are all genera of ants. 

## Question 7: What is one analysis you might perform to confirm that these BLAST hits are real? (5pts)

Check to see whether or not these results have a significant E-value. The closer the number is to 0.0 the more significant the E-Value, which can help verify that the BLAST results are accurate. 

## Question 8: Perform the analysis you suggest in Q7. (10pts) 

```python
for individual_alignment in alignment_set:
    print(individual_alignment.hsps[0].expect)
```
